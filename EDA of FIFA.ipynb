```python
# Histogram

plt.hist(df['Age'])
plt.xlabel('Players Age')
plt.ylabel('Number of players')
plt.grid(True)
plt.show()

# Players age is mostly in range from 20 to 35 
# The average age of players is 25 
# Maximum No of players are aged 24
# The youngest players age is 16 and the oldest players age is 47
```

```python
df.corr()
```

```python
#Heat map
plt.figure(figsize=(10,10))
corr=df[['Potential','Name','Age','Overall','Strength','Acceleration']].corr()
sns.heatmap(corr,annot=True)


# Potential and age are negatively correlated
# Potential and Overall are highly correlated
# Strength and age are positively correlated
```

```python
#countplot on nationality

plt.figure(figsize=(10,30))
sns.countplot(data=df,y="Nationality")
plt.grid(True)

# here we can see that most of the player are from England nationality
# country with least number of players are fiji,eriteria,oman,cuba and Honkong
```

```python
# Pie chart on preferred foot

df["Preferred Foot"].value_counts(normalize=True).plot(figsize=(8,8),kind="pie",autopct="%.2f%%")
plt.grid(True)
plt.show

#76.81% player preferred foot is Right
#23.19% player preferred foot is Left
```

```python
#subplot on Position and age

plt.figure(1)
plt.subplot(2,2,1)
df["Position"].value_counts(normalize=True).plot(figsize=(10,8),kind="bar",color="r")
plt.title("Number of player postion frequency diagram")
plt.xlabel("postion")
plt.ylabel("No. of player")

plt.subplot(2,2,2)
df["Age"].value_counts(normalize=True).plot(figsize=(12,10),kind="bar",color="g")
plt.title("Number of player Age frequency diagram")
plt.xlabel("Age")
plt.ylabel("No. of player")

# Most preferred position is ST and least preferred position is LF
# Players with age 21 are more compared to others
```

```python
# scatter plot on Work Rate and overall rating
sns.scatterplot(df["Age"],df["Overall"])
plt.title("Work Rate vs Overall ratings")
plt.grid(True)
plt.show()

# Players rated between 90-80 are mostly from 25 to 30 age bracket compared to other age group
```

```python
# Age Distribution

df=df[df['Nationality'].isin(['England', 'Germany', 'Spain'])]
sns.boxplot(x= df['Nationality'], y = df['Age'])
plt.show()
```
